```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("breastcancer.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.bare_nuclei.unique()
```

```python
def convertbare(i):
    for i in df.bare_nuclei:
        if i == "?":
            return 99999
        else :
            return i

df.bare_nuclei = df.bare_nuclei.apply(convertbare).astype("int")
```

```python
df = df.iloc[:,1:]
```

```python
df.info()
```

```python
df["class"].unique()
```

```python
df_labels = []
for i in df["class"]:
    if i == 2:
        df_labels.append([1, 0])
    elif i == 4 :
        df_labels.append([0, 1])
```

```python
df_features = df.iloc[:,:-1]
```

```python
df_labels = np.array(df_labels)
```

```python
df_labels.shape
```

```python
df_features.info()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_features, df_labels, test_size = 0.2)
```

```python
model = Sequential()
```

```python
model.add(Dense(500, input_dim = 9, activation = 'sigmoid'))
model.add(Dense(100, activation ='tanh'))
model.add(Dense(2, activation ='softmax'))
model.compile(loss ='mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
model.fit(x_train, y_train, epochs=500, batch_size=70, validation_data=(x_test, y_test))
```

```python
feature1 = np.array([[5,1,1,1,2,1,3,1,1]])
feature2 = np.array([[8,10,10,8,7,10,9,7,1]])
```

```python
predict1 = model.predict_classes(feature1)
predict2 = model.predict_classes(feature2)
```
