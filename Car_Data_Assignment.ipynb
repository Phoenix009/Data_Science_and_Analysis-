```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from keras.models import Sequential
```

```python
df = pd.read_csv("car_data.csv")
```

```python
le = preprocessing.LabelEncoder()
```

```python
le.fit(['vhigh', 'high', 'med', 'low'])
```

```python
df.buying = pd.Series(le.transform(df.buying))
```

```python
df.maint = pd.Series(le.transform(df.maint))
```

```python
le.fit(['small', 'med', 'big'])
```

```python
df.lug_boot = pd.Series(le.transform(df.lug_boot))
```

```python
le.fit(['low', 'med', 'high'])
```

```python
df.safety = pd.Series(le.transform(df.safety))
```

```python
df.head()
```

```python
df.doors.unique()
```

```python
df.persons.unique()
```

```python
def convert_doors(i):
    if i == "5more":
        return 5
    else:
        return i
```

```python
def convert_persons(i):
    if i == "more":
        return 5
    else:
        return i
```

```python
df.doors = df.doors.apply(convert_doors).astype("int")
```

```python
df.persons = df.persons.apply(convert_persons).astype("int")
```

```python
df.head()
```

```python
df.info()
```

```python
df.class_value.unique()
```

```python
df_label = []
for i in df.class_value:
    if i == "unacc":
        df_label.append([1, 0, 0, 0])
    elif i == "acc":
        df_label.append([0, 1, 0, 0])
    elif i == "vgood":
        df_label.append([0, 0, 1, 0])
    elif i == "good":
        df_label.append([0, 0, 0, 1])
```

```python
df_label = np.array(df_label)
```

```python
df_label.shape
```

```python
df_features = df.iloc[:,:-1]
```

```python
df_features.head()
```

```python
df_features.info()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_features, df_label, test_size = 0.2)
```

```python
model = Sequential()
```

```python
model.add(Dense(500, input_dim=6, activation='sigmoid'))
model.add(Dense(100, activation='tanh'))
model.add(Dense(4, activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=1000, batch_size=70, validation_data=(x_test, y_test))
```

```python
feature1 = np.array([[3, 3, 2, 2, 2, 1]])
feature2 = np.array([[1, 1, 4, 5, 0, 0]])
feature3 = np.array([[1, 1, 4, 5, 0, 2]])
feature4 = np.array([[1, 1, 5, 4, 2, 2]])
```

```python
predict1 = model.predict_classes(feature1)
predict2 = model.predict_classes(feature2)
predict3 = model.predict_classes(feature3)
predict4 = model.predict_classes(feature4)
```
